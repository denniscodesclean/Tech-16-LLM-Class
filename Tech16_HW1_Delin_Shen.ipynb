```python

from openai import OpenAI
from google.colab import userdata, files
import fitz
```

```python
import fitz
from google.colab import files

class pdf_summarizer:

  def extract_text_from_pdf(self, pdf_path):
    """
    Take PDF as input and return string.
    """
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text


  def upload_files(self):
    """
    Ask user to upload 2 pdf files.
    """
    uploaded = files.upload()

    # Ensure 2 files present
    if len(uploaded) != 2:
        raise ValueError("Please upload exactly two PDF files.")
    else:
        # Extract filenames
        pdf_files = list(uploaded.keys())

    # Extract text from each PDF
    extracted_texts = []
    for pdf_file in pdf_files:
        text = extract_text_from_pdf(pdf_file)
        extracted_texts.append(text)

    # Assign extracted texts to variables
    self.doc1, self.doc2 = extracted_texts
    return self.doc1,self.doc2

  def call_openai(self):
      # load api key
      open_ai_key = userdata.get('open_ai_api')

      # promtps
      system_prompt="""
      You will receive 2 documents.
      Response in follow steps:
      ** Context **
      explain what 2 docs talks about in general (in 20 words).
      ** Document 1 Summary **
      Summarize doc1 in 30 words.
      ** Document 2 Summary **
      Summarize doc2 in 30 words.
      ** Docs Comparison **
      Compare common and differences beteen doc1 and doc2 in 30 words.
      """

      user_prompt=f"""
      doc1:{self.doc1}
      doc2:{self.doc2}
      """

      client = OpenAI(api_key=open_ai_key)
      response = client.chat.completions.create(
          model='gpt-4o-mini',
          messages=[
              {"role":"system","content":system_prompt},
              {"role":"user","content":user_prompt}
          ]
      )

      return response.choices[0].message.content
```

```python
# Usage
summarizer = pdf_summarizer()
summarizer.upload_files()  # Upload and extract text
summary = summarizer.call_openai()  # Get AI-generated summary
print(summary)
```

**Context**  
Both documents discuss developments regarding China's DeepSeek AI company, highlighting achievements and concerns regarding potential intellectual property theft from OpenAI.

**Document 1 Summary**  
DeepSeek's advancements in AI technology have caused excitement; however, analysts express concerns over chip limitations and competitive pressures from U.S. companies.

**Document 2 Summary**  
There are allegations that DeepSeek may have copied OpenAI's ChatGPT, raising concerns about intellectual property and the implications for U.S. tech companies amidst competitive tensions.

**Docs Comparison**  
Both documents address DeepSeek's AI innovations and competitive dynamics with U.S. firms, but Document 1 focuses on achievements, while Document 2 emphasizes potential intellectual property issues.
